```
import cv2
import numpy as np
from google.colab.patches import cv2_imshow  # Use for displaying images in Colab

video_path = 'sample.mp4'
cap = cv2.VideoCapture(video_path)

mog = cv2.bgsegm.createBackgroundSubtractorMOG()

while cap.isOpened():
    ret, frame = cap.read()
    if not ret:
        break
    
    fg_mog = mog.apply(frame)
    cv2_imshow(fg_mog)
    
    if cv2.waitKey(30) & 0xFF == ord('q'):
        break

cap.release()
cv2.destroyAllWindows()
```

```
import cv2
import numpy as np
from google.colab.patches import cv2_imshow

video_path = 'sample.mp4'
cap = cv2.VideoCapture(video_path)

ret, prev_frame = cap.read()
prev_gray = cv2.cvtColor(prev_frame, cv2.COLOR_BGR2GRAY)

while cap.isOpened():
    ret, frame = cap.read()
    if not ret:
        break
    
    curr_gray = cv2.cvtColor(frame, cv2.COLOR_BGR2GRAY)
    frame_diff = cv2.absdiff(curr_gray, prev_gray)
    cv2_imshow(frame_diff)
    
    prev_gray = curr_gray
    
    if cv2.waitKey(30) & 0xFF == ord('q'):
        break

cap.release()
cv2.destroyAllWindows()
```

```
import cv2
import numpy as np
from google.colab.patches import cv2_imshow

video_path = 'sample.mp4'
cap = cv2.VideoCapture(video_path)

ret, frame = cap.read()
avg_frame = np.float32(frame)

while cap.isOpened():
    ret, frame = cap.read()
    if not ret:
        break
    
    cv2.accumulateWeighted(frame, avg_frame, 0.02)
    bg_avg = cv2.convertScaleAbs(avg_frame)
    fg_avg = cv2.absdiff(frame, bg_avg)
    
    cv2_imshow(cv2.cvtColor(fg_avg, cv2.COLOR_BGR2GRAY))
    
    if cv2.waitKey(30) & 0xFF == ord('q'):
        break

cap.release()
cv2.destroyAllWindows()
```
